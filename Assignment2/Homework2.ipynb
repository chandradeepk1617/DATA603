```python
import numpy as np
import pandas as pd
import nltk
nltk.download('words')
from nltk.corpus import words
```

### READING CONTENT FROM FILE 1

```python
with open('File1.txt','r') as file:
    Str = file.read()
print(Str)
```

### MAPPER FUNCTION FOR WORD COUNT

```python
def func_mapper(text) :
    words = text.split()
    wordCounts = [[word,1] for word in words]
    return wordCounts;
```

### REDUCER FUNCTION FOR WORD COUNT

```python
def func_reducer(counts,fileName) :
    finalCounts = dict()
    for word,count in counts :
        if word in finalCounts.keys() :
            finalCounts[word] = finalCounts[word] + 1;
        else :
            finalCounts[word] = 1
    
    import json 
    with open(fileName, 'w') as convert_file:
        convert_file.write(json.dumps(finalCounts))
    return finalCounts;
func_reducer(func_mapper(Str),"Output1.txt")
```

```python
### STORING ALL ENGLISH WORDS INTO A VARIABLE
english_vocab = list(set(words.words()))
english_vocab.append('jk')
english_vocab.append('rowling')
english_vocab.append('percy')
```

```python
##READING FILE2
with open('File2.txt','r') as file:
    Str1 = file.read()


### FUNCTION TO CHECK A WORD IS ENGLISH WORD OR NOT
def is_english_word(word):
    return word.lower() in english_vocab
englishWords = []



import re

for word in Str1.split() :
    w = re.sub(r'[^\w\s]', '', word)
    if not is_english_word(w) :
        englishWords.append(w)


func_reducer(func_mapper(" ".join(englishWords)),"Output2.txt")
```
